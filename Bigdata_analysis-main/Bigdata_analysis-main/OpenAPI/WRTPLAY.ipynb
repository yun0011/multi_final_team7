openAPI에서 "계곡" 인 부분만 따오기

```python
import requests
import pprint
import xml.etree.ElementTree as ET
import pandas as pd


url = "http://safemap.go.kr/openApiService/data/getWtrPlayData.do"
params = {
"serviceKey": "PHUYHKKL-PHUY-PHUY-PHUY-PHUYHKKLT1",
"numOfRows": "1334",
"pageNo": "1"
}

response = requests.get(url, params=params)

content = response.text

root = ET.fromstring(content)

# XML 데이터 구조 탐색
items = root.find("body").find("items")

counter = 0  # 반복 횟수를 저장할 변수

df1 = pd.DataFrame()  # 데이터를 저장할 데이터프레임 생성

for item in items.findall("item"):
    wtrplay_plc_type = item.find("WTRPLAY_PLC_TYPE").text
    if wtrplay_plc_type == "계곡":
        counter += 1  # 반복 횟수 증가
        # 추출한 데이터 처리
        wtrplay_plc_nm = item.find("WTRPLAY_PLC_NM").text
        ctprvn_nm = item.find("CTPRVN_NM").text
        sgg_nm = item.find("SGG_NM").text
        emd_nm = item.find("EMD_NM").text
        detail_plcnm = item.find("DETAIL_PLCNM").text
        adres = item.find("ADRES").text
        wtrplay_plc_type = item.find("WTRPLAY_PLC_TYPE").text
        wtrplay_sec = item.find("WTRPLAY_SEC").text
        wtrplay_deep = item.find("WTRPLAY_DEEP").text
        wtrplay_deep_avg = item.find("WTRPLAY_DEEP_AVG").text
        year_visitor = item.find("YEAR_VISITOR").text
        acdt_co = item.find("ACDT_CO").text
        x = item.find("X").text
        y = item.find("Y").text
        manage_type = item.find("MANAGEMENT_TYPE").text
        manage_at = item.find("MANAGE_AT").text

        #추출한 데이터 데이터프레임으로 변경
        df = pd.DataFrame( 
           [[wtrplay_plc_nm, ctprvn_nm, sgg_nm, emd_nm, detail_plcnm, adres, wtrplay_plc_type, wtrplay_sec, wtrplay_deep, wtrplay_deep_avg, year_visitor, acdt_co, x, y, manage_type, manage_at]],

            columns=["wtrplay_plc_nm", "ctprvn_nm", "sgg_nm", "emd_nm", "detail_plcnm", "adres", "wtrplay_plc_type", "wtrplay_sec", "wtrplay_deep", "wtrplay_deep_avg", "year_visitor", "acdt_co", "x", "y", "manage_type", "manage_at"]
        )

        # 데이터프레임에 추출한 데이터 추가
        df1 = pd.concat([df1, df])

print(df1)

print("반복 횟수:", counter)

df1.to_csv("WRTPLAY.csv", index=False, encoding="cp949")
```
